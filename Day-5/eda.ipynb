```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# make plots look good
sns.set(style="whitegrid")

# LOAD DATASET
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python

# Basic info
df.info()
```

- Total rows
- Data types
- Missing values

```python

# Statistical summary
df.describe()
```

- Mean age
- Max fare
- Distribution

```python

# Target variable check
df['Survived'].value_counts()
```

- How many survived vs died

```python

# MISSING VALUES
df.isnull().sum()
```

- Age has missing values
- Cabin has many missing values
- Embarked has few missing values

```python

# DATA CLEANING
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

- Cabin can be ignored (too many missing)

```python

# age distribution
plt.figure(figsize=(6,4))
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.show()
```

- Most passengers are between 20–40

```python

# Survival count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

- Females survived more than males

```python

# Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Gender vs Survival")
plt.show()
```

- Females survived more than males

```python

# Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Class vs Survival")
plt.show()
```

- 1st class survived more

```python

# CORRELATION
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

- Fare & Pclass related
- Some weak correlations

```python

#PAIRPLOT (ADVANCED)
sns.pairplot(df[['Survived','Age','Fare','Pclass']])
```

Key Insights:

- Female passengers had higher survival rate
- Passengers in 1st class survived more
- Age distribution shows majority between 20–40
- Fare is positively related to class
- Survival depends on gender and class strongly
